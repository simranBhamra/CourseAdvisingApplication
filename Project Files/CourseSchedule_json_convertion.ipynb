```python
mystr = df.iloc[609]['MEETING INFORMATION']
```

```python
def split_info(meeting_info):
    try:
        info_split = str(meeting_info).split(' ')
        #print(info_split)
        start_date = info_split[0]
        end_date = info_split[1]
        days = ''
        start_time = ''
        if info_split[-2][:3] == 'MWF':
            days = info_split[-2][:3]
            start_time = info_split[-2][3:]
        else:
            days = info_split[-3]
            start_time = info_split[-2]
        end_time = info_split[-1]

        return (start_date,end_date,days,start_time,end_time)
    except:
        return ('','','','','')
split_info(mystr)
```

```python
df[['start_date','end_date','days','start_time','end_time']] = df['MEETING INFORMATION'].apply(split_info).apply(pd.Series)
```

```python
mylist = []

for i in range(len(df)-1):
    row = df.iloc[i]
    title = row['TITLE']
    course = row['COURSE']
    startDate = row['start_date']
    endDate = row['end_date']
    startTime = row['start_time']
    endTime = row['end_time']
    classDays = row ['days']
    
    
    class_dict = {'title':row['TITLE'],
                 'course':row['COURSE'],
                 'startDate':row['start_date'],
                 'endDate':row['end_date'],
                 'startTime':row['start_time'],
                 'endTime':row['end_time'],
                 'days':row['days']}
   
    mylist.append(class_dict)
```
