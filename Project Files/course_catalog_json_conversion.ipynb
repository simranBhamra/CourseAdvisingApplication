```python
courses = df['COURSE'].to_list()

course_prereqs = []

for i in range(len(df)-1):
    row = df.iloc[i]
    #print(row['COURSE'])
    prerequisites_string = str(row['PREREQUISITES'])
    prerequisites_list = []
    for course_sub in courses:
        #print('course_sub' + course_sub)
        if course_sub.lower() in prerequisites_string.lower():
            prerequisites_list.append(course_sub)
    #print(prerequisites_list)
    current_prereq_dict = {'course':row['COURSE'],'title':row['COURSE TITLE'],'prerequisites':prerequisites_list}
    #print(current_prereq_dict['prerequisites'])
    course_prereqs.append(current_prereq_dict)
```

```python
outdict = {'courses':course_prereqs}

import json

print(json.dumps(outdict, indent=4, sort_keys=True))


with open('courses.json', 'w', encoding='utf-8') as f:
    json.dump(outdict, f, ensure_ascii=False, indent=4)
```
